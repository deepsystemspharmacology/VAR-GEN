```python
import pandas as pd
import collections
import numpy as np


orig_file = "raw_files/featurecounts.genes.TPM.selected_reps.ln.surrogate_variables_corrected.protein_coding.txt"
```

```python
orig_df = pd.read_csv(orig_file, index_col=0, sep="\t")

print "Original data"
print orig_df.shape
print orig_df.columns.values
print orig_df.head()
```

```python
rep2cols = collections.OrderedDict()
ts = []


for col in orig_df.columns.values:
    t = col.split("_")[0]
    rep = col.split("_")[1][:4]
    
    print col, t, rep
    
    if t not in ts:
        ts.append(t)
        
    if rep not in rep2cols:
        rep2cols[rep] = []
    
    if col not in rep2cols[rep]:
        rep2cols[rep].append(col)
    
reps = rep2cols.keys()
repcol_lists = rep2cols.values()

for rep, repcol_list in zip(reps, repcol_lists):
    print rep
    print "# cols: ", len(repcol_list)
    print repcol_list
    
        
    
```

```python
rep_dfs = []

for rep, repcol_list in zip(reps, repcol_lists):
    rep_df = orig_df[repcol_list]
    rep_dfs.append(rep_df)
    
    print rep
    print rep_df.shape
    print rep_df.head()
```

### Perform imputation via linear combination of nearest replciates
Weighted by distance

Doesn't work for both endpoints

```python
suffix = "impute-linear"
imputed_rep_dfs = []

print suffix

for rep, rep_df in zip(reps, rep_dfs):
    imputed_rep_df = rep_df.copy()

    orig_rep_cols = rep_df.columns.values
    orig_rep_ts = [x.split("_")[0] for x in orig_rep_cols]
    missing_ts = set(ts).difference(set(orig_rep_ts))

    print rep
    print "Missing timepoints: ", missing_ts
    
    for mt in missing_ts:
        m_index = ts.index(mt)
        
        left_index = m_index
        left_step = 0
        while ts[left_index] not in orig_rep_ts:
            left_step += 1
            left_index = m_index - left_step
            
        right_index = m_index
        right_step = 0
        while ts[right_index] not in orig_rep_ts:
            right_step += 1
            right_index = m_index + right_step
        
        left_t = ts[left_index]
        left_col = [x for x in rep_df.columns.values  if x.split("_")[0] == left_t and x in orig_rep_cols][-1]
        right_t = ts[right_index]
        right_col = [x for x in rep_df.columns.values if x.split("_")[0] == right_t and x in orig_rep_cols][0]
        new_m_col = mt + "_" + rep + "_" + suffix
        
        # the new column should be the weightedl inear combination of the nearest endpoints
        imputed_rep_df[new_m_col] = (imputed_rep_df[left_col] * right_step + imputed_rep_df[right_col] * left_step) * 1.0/(right_step + left_step)

        
        
        print "Imputing ", mt, "to", new_m_col
        print "Left col", left_col
        print "Right col", right_col
        
        print imputed_rep_df[[left_col, new_m_col, right_col]].head()
    
    
        # Sort by the order in the timepoints
        ordered_cols = sorted(list(imputed_rep_df.columns.values), key=lambda col: ts.index(col.split("_")[0]))
        print "Ordered cols", ordered_cols
        imputed_rep_df = imputed_rep_df[ordered_cols]
    
    imputed_rep_dfs.append(imputed_rep_df)
        
            
        
        
```

### Save the new files

```python
print "Impute suffix is ", suffix

outfiles = []

for rep, imputed_rep_df in zip(reps, imputed_rep_dfs):
    is_imputed = any([x.endswith(suffix) for x in imputed_rep_df.columns.values])
    
    outfile = orig_file.replace("reps", rep if not is_imputed else rep + "-" + suffix)
    outfiles.append(outfile)
    
    imputed_rep_df.to_csv(outfile, sep="\t")
    
    print rep
    print "Imputed: ", is_imputed
    print "Writing to ", outfile
    
    
    
data_list_file = "raw_files/rep_file_list.txt"
data_list_dict = collections.OrderedDict()
data_list_dict["Rep"] = reps
data_list_dict["File"] = outfiles
data_list_df = pd.DataFrame(data_list_dict)
data_list_df.to_csv(data_list_file, sep="\t", index=False)

# with open(data_list_file, 'w') as dfile:
#     dfile.write("Rep File")
#     for rep, outfile in zip(rep, outfiles):
#         dfile.write(rep + "\t" + outfile + "\n")

print "List of  data written to ", data_list_file
```
