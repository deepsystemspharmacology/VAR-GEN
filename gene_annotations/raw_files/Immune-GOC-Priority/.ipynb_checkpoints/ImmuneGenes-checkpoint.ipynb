```python
import collections
import csv 
raw_file = "ImmuneHGNCGenes-first-scrape.txt"
new_file = "ImmuneHGNCGenes-second-scrape.txt"
conv_file = "HGNC-to-Ensembl.txt"
new_conv_file = "HGNC-to-Ensembl-all.txt"
out_file = "ImmuneEnsembleGenes-second-scrape.txt"
final_out_file = "ImmuneEnsembleGenes-final.txt"

priority_dict = collections.OrderedDict()
priority_prefix = "Genes with Priority Score of "
gene_prefix = ">"
score = -1
with open(raw_file, 'rU') as f:
    for line in f.readlines():
        if priority_prefix in line:
            score = line.split(priority_prefix)[1][:-1]
            print "Score: ", score
            if score not in priority_dict:
                priority_dict[score] = set()
        elif gene_prefix in line:
            gene = line.split(gene_prefix)[1][:-1]
            print "Gene: ", gene
            priority_dict[score].add(gene)
```

```python
with open(new_file, 'w') as n:
    writer = csv.writer(n, delimiter="\t")
    writer.writerow(["Gene", "Priority"])
    for priority in priority_dict:
        for gene in priority_dict[priority]:
            writer.writerow([gene, priority])
            
!head $new_file
```

```python
import csv
ensembl_col = "Ensembl Gene ID"
conv_dict = {}
with open(conv_file, 'rU') as csvfile:
    reader = csv.DictReader(csvfile, delimiter="\t")
    conv_fields = reader.fieldnames[:]
    conv_fields.remove(ensembl_col)
    for row in reader:
        for conv_field in conv_fields:
            entry = row[conv_field]
            all_keys = entry.split(",")
            for key in all_keys:
                new_key = key[:]
                new_key = new_key.replace(" ", "")
                conv_dict[new_key] = row[ensembl_col]

for item in conv_dict.items()[0:30]:
    print item
    
```

```python
with open(new_conv_file, 'w') as f:
    writer = csv.writer(f, delimiter="\t")
    writer.writerow(["HGNC-all", "Ensembl Gene ID"])
    for item in conv_dict.items():
        writer.writerow(item)
```
